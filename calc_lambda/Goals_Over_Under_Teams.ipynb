```python
import pandas as pd
import numpy as np
```

```python
# we'll take lambda values from this table
lambda_table = pd.read_csv('lambda.csv', sep = ';', index_col=0)
lambda_table
```

```python
def calc_lambda(line_h, line_a, u_h, u_a, o_h, o_a, goals_scored_h = 0, goals_scored_a = 0):
    """
    Function returns lambda depending on inputs
    Variables:
    line_h - num of goals in line for home
    line_a - num of goals in line for away
    u_h - under home coef
    u_a - under away coef
    o_h - over home coef
    o_a - over away coef
    goals_scored_h - goals scored by home team
    goals_scored_a - goals scored by away team
    """
    
    k_h = round(1/(u_h*(1/u_h + 1/o_h)), 2)
    k_a = round(1/(u_a*(1/u_a + 1/o_a)), 2)
    
    # update line according to current score
    line_h -= goals_scored_h # shouldn't be < 0
    line_a -= goals_scored_a # shouldn't < 0
    if (line_h or line_a) < 0:
        return ('wrong line/goal inputs')
    
    # finding corresponding lanbda from table where row = k, column = line
    # need to handle situation when there is no corresponding lambda value
    lambda_h = round(lambda_table.loc[k_h, str(line_h)],4)
    lambda_a = round(lambda_table.loc[k_a, str(line_a)],4)
    
    print('k_h = {} \nk_a = {}'.format(k_h, k_a))
    print('line_h = {} \nline_a = {}'.format(line_h, line_a))
    print('lambda_h = {} \nlambda_a = {}'.format(lambda_h, lambda_a))
    
    return [lambda_h, lambda_a]
```

```python
calc_lambda(line_h=1.5, line_a=1.5, u_h =1.4, u_a=3, o_h=2.2, o_a=1.3)
```

```python
calc_lambda(line_a=2.5, line_h=2.5, u_h =1.4, u_a=3, o_h=2.2, o_a=1.3)
```
